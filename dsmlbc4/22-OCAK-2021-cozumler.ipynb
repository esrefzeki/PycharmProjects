```python
a = dt.datetime.strptime(tarih, "%Y-%m-%d").weekday()
```

```python
dayName = ["Pazar", "Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma", "Cumartesi"]
dayName[a]
```

```python
def get_day(str_date, date_format="%Y-%m-%d"):
    import datetime
    dayName = ["Pazar", "Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma", "Cumartesi"]
    day = dt.datetime.strptime(str_date, date_format).weekday()
    return dayName[day]
```

```python
df["DAY"] = df["DATE_2"].apply(lambda x: get_day(x.split()[0], "%Y-%m-%d"))
```

```python
df["DAY"].head()
```

```python
df.groupby("DAY").agg({"LINENETTOTAL": "sum"}).sort_values("LINENETTOTAL", ascending=False)
```
